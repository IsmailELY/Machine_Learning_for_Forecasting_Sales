```python
def epoch_converter(date):
    return dtm.datetime(int(date.split("-")[0]),int(date.split("-")[1]),int(date.split("-")[2]),0,0).timestamp()
def long_lat_region(region):
    value = reg.index(region)
    return long[value],lat[value]
```

```python
Region = myData[["REGION NAME","GREGORIAN_DATE"]].drop_duplicates()
Region["Epoch_time"] = Region["GREGORIAN_DATE"].map(lambda x: epoch_converter(x))
Region["Longetude"] = myData["REGION NAME"].map(lambda x:long_lat_region(x)[0])
Region["Latitude"] = myData["REGION NAME"].map(lambda x:long_lat_region(x)[1])
```
