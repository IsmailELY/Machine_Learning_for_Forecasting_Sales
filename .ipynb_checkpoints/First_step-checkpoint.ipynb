<h3>1- Holiday Column</h3>

```python
Dates = myData["PERIODACTIVEDAY"]
myData["GREGORIAN_DATE"] = myData["PERIODACTIVEDAY"].map(lambda x: x[:-9])
myData=myData.drop("PERIODACTIVEDAY",axis=1)
myData.head()
```

```python
alger_holi = pd.read_csv("algerian_holiday.csv",header=0,encoding="latin-1")
scolar_holi = pd.read_csv("scolar_holidays.csv",header=0,encoding="utf-16")
hijri_events = pd.read_csv("hijri_holidays.csv",header=0,encoding="latin-1")
hijri_events.head()
```

```python
def holiday(date,dates):
    if date in dates:
        return 1
    else:
        return 0
algerH_serie = myData["GREGORIAN_DATE"].map(lambda x: holiday(x,alger_holi["GREGORIAN_DATE"].tolist()))
algerH_serie.head(),algerH_serie.size
```

```python
scolarH_serie = myData["GREGORIAN_DATE"].map(lambda x: holiday(x,scolar_holi["GREGORIAN_DATE"].tolist()))
scolarH_serie.head(),scolarH_serie.size
```

```python
def hijri_holiday(date,dates,event):
    for i in range(len(dates)):
        if date==dates[i]:
            return event[i]
        else:
            return "None"
hijri_serie = myData["GREGORIAN_DATE"].map(lambda x: hijri_holiday(x,hijri_events["GREGORIAN_DATE"].tolist(),hijri_events["ISLAMIC_HOLIDAY"].tolist()))
hijri_serie.head(),hijri_serie.size
```

```python
myData["HOLIDAY"] = algerH_serie
myData["SCOLAR_HOLIDAY"] = scolarH_serie
myData["ISLAM_EVENT"] = hijri_serie
myData.head()
```

<h3>2 - Heat and temperature </h3>

<p>Isolate the date & regions column to minimize the work</p>

```python
import datetime as dtm
import requests as rq

details = pd.read_csv("coord.csv",header=0)
reg = details["REGION"].tolist()
long = details["LONGETUDE"].tolist()
lat = details["LATITUDE"].tolist()
API_Keys = ["48aa4110997c86a753f57dbdf9e4c63d"]

def epoch_converter(date):
    return dtm.datetime(int(date.split("-")[0]),int(date.split("-")[1]),int(date.split("-")[2]),0,0).timestamp()
def long_lat_region(region):
    value = reg.index(region)
    return long[value],lat[value]

def get_temperature(long,lat,time):
    response = None
    while response is None:
        key=API_Keys[0]
        print("https://api.darksky.net/forecast/"+key+"/"+str(long)+","+str(lat)+","+str(int(time)))
        resp = rq.get("https://api.darksky.net/forecast/"+key+"/"+str(long)+","+str(lat)+","+str(int(time)))
        response = resp.json()
        try: 
            response = response["currently"]
        except KeyError as e:
            response=None
            API_Keys.pop(0)
    temp = None
    hum = None
    try :
        temp=response["temperature"]
    except KeyError as e:
        pass
    try :
        hum = response["humidity"]
    except KeyError as e:
        pass
    return temp,hum
```

```python
Region = myData[["REGION NAME","GREGORIAN_DATE"]].drop_duplicates()
Region["Epoch_time"] = Region["GREGORIAN_DATE"].map(lambda x: epoch_converter(x))
Region["Longetude"] = myData["REGION NAME"].map(lambda x:long_lat_region(x)[0])
Region["Latitude"] = myData["REGION NAME"].map(lambda x:long_lat_region(x)[1])
```

```python
Region.to_csv("Alger_Region_Upgraded.csv",index=False)
```

```python
RegionR = pd.read_csv("Alger_Region_Upgraded.csv",header = 0)
```

```python
API_Dict = {"temperature":[],"humidity":[]}
for i in range(19884,RegionR.index.max()+1):
    epoch=RegionR["Epoch_time"][i]
    long =RegionR["Longetude"][i]
    lat  =RegionR["Latitude"][i]
    resp =get_temperature(long,lat,epoch)
    API_Dict["temperature"].append(resp[0])
    API_Dict["humidity"].append(resp[1])
```
